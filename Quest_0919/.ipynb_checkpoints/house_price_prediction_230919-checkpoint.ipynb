## 데이터 모델링

````python
# RMSE 손실함수 구현
def rmse(y_test, y_pred):
    return np.sqrt(mean_squared_error(np.expm1(y_test), np.expm1(y_pred)))

# 모델별 학습 및 예측값 평가
def get_scores(models, train, y):
    df = {}
    for model in models:
        # 모델 이름
        model_name = model.__class__.__name__
        # 학습/테스트 데이터 분리
        X_train, X_test, y_train, y_test = train_test_split(train, y, test_size=0.2, random_state=random_state)
        # 모델 학습
        model.fit(X_train, y_train)
        # 결과 예측
        y_pred = model.predict(X_test)
        # RMSE값 저장
        df[model_name] = rmse(y_test, y_pred)
        score_df = pd.DataFrame(df, index=['RMSE']).T.sort_values('RMSE', ascending=False)
    return score_df
````

````python
# 랜덤 시드 고정
random_state = 42

# 모델 생성
gboost = GradientBoostingRegressor(random_state=random_state)
xgboost = XGBRegressor(random_state=random_state)
lightgbm = LGBMRegressor(random_state=random_state)
rdforest = RandomForestRegressor(random_state=random_state)
models = [gboost, xgboost, lightgbm, rdforest]

# 모델별 학습 및 예측값 평가
get_scores(models, train, y)
````

> 모델을 평가해 보니, LGBMRegressor 모델이 가장 손실율이 적은 것으로 확인됨

````python
# 파라미터 최적 값을 찾기 위한 그리드 서치 함수 정의
def my_GridSearch(model, train, y, param_grid, verbose=2, n_jobs=5):
    # 1. GridSearchCV 모델로 `model`을 초기화
    grid_model = GridSearchCV(model, param_grid=param_grid, \
                        scoring='neg_mean_squared_error', \
                        cv=5, verbose=1, n_jobs=5)
    #2. 모델을 fitting
    grid_model.fit(train, y)
    # 3. params, score에 각 조합에 대한 결과를 저장
    params = grid_model.cv_results_['params']
    score = grid_model.cv_results_['mean_test_score']
    # 4. 데이터 프레임을 생성하고, RMSLE 값을 추가한 후 점수가 높은 순서로 정렬
    results = pd.DataFrame(params)
    results['score'] = score
    results['RMSLE'] = np.sqrt(-1 * results['score'])
    results = results.sort_values('RMSLE', ascending=False)
    # 5. results 반환
    return results
````

````python
# prams 정의
param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [1, 10],
}
````

````python
# LGBMRegressor모델에 대한 그리드 서치 실행
my_GridSearch(LGBMRegressor(random_state=random_state), train, y, param_grid)
````

````python
# RandomForestRegressor 모델에 대한 그리드 서치 실행
my_GridSearch(RandomForestRegressor(random_state=random_state), train, y, param_grid)
````

````python
# GradientBoostingRegressor 모델에 대한 그리드 서치 실행
my_GridSearch(GradientBoostingRegressor(random_state=random_state), train, y, param_grid)
````

* 모델 앙상블

> 위 과정에서 찾은 각 모델 함수의 최적의 하이퍼 파라미터 값들을 사용함

### 앙상블 함수 StackingRegressor()

`StackingRegressor`는 scikit-learn 라이브러리에서 제공하는 앙상블 학습 방법 중 하나입니다. 앙상블 학습은 여러 다른 모델을 조합하여 단일 예측 모델을 만들어 더 좋은 성능을 얻기 위한 기술입니다. `StackingRegressor`는 회귀(Regression) 문제에 대한 앙상블 모델을 만드는 데 사용됩니다.  

여기서 `StackingRegressor` 함수의 매개변수를 해석해보겠습니다:  

* `stack_models`: 이 매개변수는 리스트 형태로 제공되며, 앙상블 스택을 구성할 기본 회귀 모델들의 리스트입니다. 예를 들어, 다양한 회귀 모델(예: 선형 회귀, 결정 트리 회귀, 랜덤 포레스트 회귀 등)을 생성하고 이를 리스트에 추가하여 앙상블 스택을 구성할 수 있습니다. 앙상블 스택은 이러한 기본 모델들의 예측 결과를 바탕으로 최종 예측을 수행합니다.  

* `final_estimator`: 이 매개변수는 앙상블 스택의 최종 예측 모델을 지정합니다. 여기서 `gboost`는 그레디언트 부스팅 회귀(Gradient Boosting Regression) 모델을 나타냅니다. 그레디언트 부스팅은 앙상블 학습 알고리즘 중 하나로, 이전 모델의 오차를 보완하는 방식으로 순차적으로 모델을 학습합니다. `final_estimator`로 그레디언트 부스팅 회귀 모델을 선택하면, 앙상블 스택의 최종 예측을 수행하기 위해 그레디언트 부스팅 모델을 사용합니다.  

따라서, `StackingRegressor`를 사용하면 다양한 회귀 모델을 조합하여 앙상블 스택을 생성하고, 이 스택의 최종 예측 모델로 그레디언트 부스팅 회귀 모델을 사용하여 회귀 문제를 해결할 수 있습니다. 이렇게 함으로써 모델의 성능을 향상시키고 안정성을 높일 수 있습니다.  

````python
from sklearn.ensemble import StackingRegressor

def get_stacking_model():
    gboost = GradientBoostingRegressor(max_depth=10, n_estimators=100, random_state=random_state)
    xgboost = XGBRegressor(max_depth=10, n_estimators=50, random_state=random_state)
    lightgbm = LGBMRegressor(max_depth=10, n_estimators=100, random_state=random_state)
    rdforest = RandomForestRegressor(max_depth=10, n_estimators=100, random_state=random_state)

    stack_models = [
        ('gb', gboost),
        ('xgb', xgboost),
        ('lgbm', lightgbm),
        ('rf', rdforest)
    ]
    
    # 학습/테스트 데이터 분리
    X_train, X_test, y_train, y_test = train_test_split(train, y, test_size=0.2, random_state=random_state)

    # 모델 앙상블
    stack_reg = StackingRegressor(stack_models, final_estimator=gboost)
    stack_reg.fit(X_train, y_train)

    y_pred = stack_reg.predict(X_test)
    print(f'rmse: {rmse(y_pred, y_test)}')

    return stack_reg

s_model = get_stacking_model()
````

````python
# 앙상블 모델에 대한 예측 및 예측결과를 submission.csv 파일로 저장
def save_submission(model):
    # test에 대해 예측
    prediction = model.predict(test)
    prediction = np.expm1(prediction)
    sub = pd.DataFrame(data={'id':sub_id,'price':prediction})
    # csv 파일을 저장
    sub.to_csv('submission.csv', index=False)
    print('submission complete!')    
    
save_submission(s_model)
````
